## 3. Conceptos iniciales

### Definiciones básicas
- **Clase:** el molde o definición general.
- **Objeto:** una instancia concreta creada a partir de una clase.

- **Interfaz (idea inicial):** el conjunto de acciones con las que se puede interactuar con el objeto.

## 4. Primera estructura de una clase

Observa esta estructura base:

### Respuesta

2. Es el constructor. Se ejecuta automáticamente al crear un objeto para asignar los valores iniciales a sus atributos (nombre, capacidad, etc.).

3. Representan los atributos de la instancia actual. El `self` asegura que cada objeto tenga sus propios datos independientes.

## 5. Completa la clase

Ahora agrega métodos simples a la clase.

### Tareas
- Crear un método `mostrar_info()` que imprima la información principal de la sala.
- Crear un método `reservar()` que deje la sala como no disponible si estaba libre.
- Crear un método `liberar()` que deje la sala disponible nuevamente.

### Pista
Recuerda que dentro de los métodos puedes acceder a los atributos usando `self`.


```python
class Sala:
    def __init__(self, nombre, capacidad, ubicacion, disponible=True):
        self.nombre = nombre
        self.capacidad = capacidad
        self.ubicacion = ubicacion
        self.disponible = disponible

    # Completa este método
    def mostrar_info(self):
        print(f"Nombre: {self.nombre}")
        print(f"Capacidad: {self.capacidad}")
        print(f"Ubicación: {self.ubicacion}")
        print(f"Disponible: {'Sí' if self.disponible else 'No'}")
        return

    # Completa este método
    def reservar(self):
        if self.disponible:
            self.disponible = False
            print(f"La sala '{self.nombre}' ha sido reservada.")
        else:
            print(f"La sala '{self.nombre}' tiene espacios disponibles")
        return

    # Completa este método
    def liberar(self):
        if not self.disponible:
            self.disponible = True
            print(f"La sala '{self.nombre}' ha sido liberada.")
        else:
            print(f"La sala '{self.nombre}' ya se encontraba disponible.")
        return
```

## 9. Desafío opcional

Si terminas antes, intenta una o más de estas mejoras:

- agregar el atributo `tipo_sala`
- validar que la capacidad no sea negativa
- mostrar la información en un formato más claro
- crear una lista con varias salas y recorrerla para imprimir sus datos
- impedir que una sala ya reservada se vuelva a reservar sin aviso


```python
class Sala:
    def __init__(self, nombre, capacidad, ubicacion, tipo_sala, disponible=True):
        self.nombre = nombre
        self.capacidad = max(0, capacidad)
        self.ubicacion = ubicacion
        self.tipo_sala = tipo_sala
        self.disponible = disponible

    def mostrar_info(self):
        estado = "Disponible" if self.disponible else "Reservada"
        print("="*30)
        print(f"Sala: {self.nombre} ({self.tipo_sala})")
        print(f"Ubicación: {self.ubicacion}")
        print(f"Capacidad: {self.capacidad} personas")
        print(f"Estado: {estado}")
        print("="*30)

    def reservar(self):
        if self.disponible:
            self.disponible = False
            print(f"Éxito: La sala '{self.nombre}' ha sido reservada.")
        else:
            print(f"Error: La sala '{self.nombre}' ya se encuentra ocupada.")

    def liberar(self):
        if not self.disponible:
            self.disponible = True
            print(f"Éxito: La sala '{self.nombre}' ahora está libre.")
        else:
            print(f"Aviso: La sala '{self.nombre}' ya estaba disponible.")

salas = [
    Sala("B-102", 30, "Piso 1", "Computación"),
    Sala("C-201", 50, "Piso 2", "Teoría"),
    Sala("Auditorio", 200, "Planta Baja", "Eventos")
]

# 1. Reservar una sala y tratar de reservarla de nuevo
salas[0].reservar()
salas[0].reservar()

# 2. Mostrar todas las salas
for s in salas:
    s.mostrar_info()
```
